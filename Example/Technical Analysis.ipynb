```python
from tehran_stocks import Stocks, db
import ta
import matplotlib.pyplot as plt
```

```python
stock = Stocks.query.filter_by(name='بترانس').first() #find by symbol(نماد)
print(Stocks.summary(stock))
```

```python
print(stock)
```

```python
df = ta.add_trend_ta(stock.df,'high','low','close')
```

# RSI

```python
ta.momentum.rsi(df.close).plot()
plt.axhline(30,ls='--',c='g')
plt.axhline(70,ls='--',c='r')
```

# MACD

```python
close = df.close.iloc[-300:]
ta.trend.macd_diff(close).plot(label='diff')
ta.trend.macd_signal(close).plot(label='signal')
# ta.trend.macd().macd(colse).plot(ls='--',label= 'MACD')
plt.axhline(0,ls='--',c='red')
plt.legend()
```
